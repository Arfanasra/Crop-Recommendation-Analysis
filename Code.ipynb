Aggregate Nitrogen, Phosphorus and potassium aka NPK by crop.

```python
# Group by the 'Crop' column and calculate the average Nitrogen, Phosphorus, and Potassium for each crop
average_npk = df.groupby('Crop').agg({'Nitrogen': 'mean', 'Phosphorus': 'mean', 'Potassium': 'mean'})

# Reset index to convert GroupBy object to DataFrame
average_npk_df = average_npk.reset_index()

# Display the DataFrame with only 'Crop', 'Nitrogen', 'Phosphorus', and 'Potassium' columns
average_npk_df[['Crop', 'Nitrogen', 'Phosphorus', 'Potassium']]
```

Convert the data into NPK ratio

```python
# Copy the DataFrame to preserve the original data
ratio_df = average_npk_df.copy()

# Calculate the sum of Nitrogen, Phosphorus, and Potassium for each crop
sum_npk = ratio_df[['Nitrogen', 'Phosphorus', 'Potassium']].sum(axis=1)

# Divide each nutrient value by the sum to get the ratio
ratio_df['Nitrogen'] /= sum_npk
ratio_df['Phosphorus'] /= sum_npk
ratio_df['Potassium'] /= sum_npk

# Display the ratio DataFrame
ratio_df
```

Understanding NPK

Nitrogen (N): Essential for leaf growth and overall plant vigor. It is a key component of chlorophyll, the compound that plants use in photosynthesis to convert sunlight into energy.
Phosphorus (P): Important for root development, flowering, and fruiting. It plays a crucial role in energy transfer within the plant and is vital during the early stages of growth.
Potassium (K): Enhances disease resistance, improves fruit quality, and aids in water regulation and enzyme activation within the plant.

Crop-Specific NPK Ratios
Different crops have different growth patterns, nutrient needs, and stages of development, which is why their preferred NPK ratios vary. Here are some examples:

1. Leafy Greens (e.g., Lettuce, Spinach)
Preferred NPK Ratio: High in Nitrogen
Example Ratio: 5-3-3 or 10-5-5
Reason: Leafy greens require more nitrogen to support vigorous leaf growth and lush foliage.

2. Root Vegetables (e.g., Carrots, Beets)
Preferred NPK Ratio: Balanced or slightly higher in Phosphorus
Example Ratio: 4-10-10 or 5-10-10
Reason: Phosphorus is crucial for root development, and balanced levels of nitrogen and potassium support overall growth and health.

3. Fruiting Vegetables (e.g., Tomatoes, Peppers)
Preferred NPK Ratio: Higher in Phosphorus and Potassium
Example Ratio: 5-10-10 or 8-16-16
Reason: Phosphorus promotes flowering and fruit set, while potassium enhances fruit quality and disease resistance.

4. Grains and Cereals (e.g., Wheat, Rice)
Preferred NPK Ratio: High in Nitrogen with balanced Phosphorus and Potassium
Example Ratio: 10-10-10 or 20-10-10
Reason: Nitrogen supports the vegetative growth phase, while balanced phosphorus and potassium ensure good root development and overall health.

5. Legumes (e.g., Beans, Peas)
Preferred NPK Ratio: Moderate in Nitrogen, Higher in Phosphorus and Potassium
Example Ratio: 5-10-10 or 6-24-24
Reason: While legumes fix their own nitrogen, they benefit from additional phosphorus and potassium for root growth and pod development.

Factors Influencing NPK Needs
Soil Type: The existing nutrient content of the soil can affect the NPK requirements. Soil tests are often conducted to determine the precise needs.
Growth Stage: Different stages of growth (e.g., seedling, vegetative, flowering, fruiting) may require adjustments in the NPK ratio.
Climate: Local climate conditions, including temperature and rainfall, can influence nutrient uptake and the overall fertility management strategy.

Pairing crop with temperature and humidity 

```python
import matplotlib.pyplot as plt

# Define the tolerance range for temperature and humidity
temperature_tolerance = 2  # in degrees Celsius
humidity_tolerance = 5  # in percentage points

# Initialize lists to store suitable crop pairs and their coordinates
suitable_crop_pairs = []
x = []
y = []

# Iterate through each pair of crops
for i in range(len(average_stats_df)):
    for j in range(i + 1, len(average_stats_df)):
        crop1 = average_stats_df.iloc[i]
        crop2 = average_stats_df.iloc[j]
        
        # Check if the temperature and humidity requirements of both crops fall within the tolerance range
        if (abs(crop1['Average Temperature'] - crop2['Average Temperature']) <= temperature_tolerance and
            abs(crop1['Average Humidity'] - crop2['Average Humidity']) <= humidity_tolerance):
            
            # Add the pair of crops to the list of suitable crop pairs
            suitable_crop_pairs.append((crop1['Crop'], crop2['Crop']))
            
            # Append the coordinates of the crops to the lists
            x.append(crop1['Average Temperature'])
            x.append(crop2['Average Temperature'])
            y.append(crop1['Average Humidity'])
            y.append(crop2['Average Humidity'])

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(average_stats_df['Average Temperature'], average_stats_df['Average Humidity'], label='Crops', color='blue')
plt.scatter(x, y, color='red', marker='x', label='Suitable Pairs')

# Connect the points of suitable crop pairs with lines
for pair in suitable_crop_pairs:
    crop1 = average_stats_df.loc[average_stats_df['Crop'] == pair[0]]
    crop2 = average_stats_df.loc[average_stats_df['Crop'] == pair[1]]
    plt.plot([crop1['Average Temperature'].values[0], crop2['Average Temperature'].values[0]], 
             [crop1['Average Humidity'].values[0], crop2['Average Humidity'].values[0]], 
             color='green', linestyle='--')

# Add labels to the points
for index, row in average_stats_df.iterrows():
    plt.annotate(row['Crop'], (row['Average Temperature'], row['Average Humidity']), textcoords="offset points", xytext=(0,10), ha='center')

# Add labels and title
plt.xlabel('Average Temperature (C)')
plt.ylabel('Average Humidity (%)')
plt.title('Crop Suitability based on Temperature and Humidity')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

Import necessary library and create heatmap for Plant Suitability Heatmap based on NPK Ratios

```python
import seaborn as sns
import numpy as np

# Extract the list of crop names from the DataFrame
plants = ratio_df['Crop'].tolist()

# Initialize an empty suitability matrix
suitability_matrix = np.zeros((len(plants), len(plants)))

# Calculate suitability between NPK ratios for each pair of plants
for i, plant1 in enumerate(plants):
    for j, plant2 in enumerate(plants):
        if i != j:
            # Calculate suitability between NPK ratios
            npk_ratio_plant1 = ratio_df[ratio_df['Crop'] == plant1][['Nitrogen', 'Phosphorus', 'Potassium']].values[0]
            npk_ratio_plant2 = ratio_df[ratio_df['Crop'] == plant2][['Nitrogen', 'Phosphorus', 'Potassium']].values[0]
            suitability = sum(abs(x - y) for x, y in zip(npk_ratio_plant1, npk_ratio_plant2)) / 3  # Average absolute difference
            suitability_matrix[i, j] = suitability

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(suitability_matrix, cmap='YlGnBu', xticklabels=plants, yticklabels=plants, annot=True, fmt=".2f")

# Add a title
plt.title('Plant Suitability Heatmap based on NPK Ratios', fontsize=14)

# Rotate tick labels for better readability
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Display the heatmap
plt.tight_layout()
plt.show()
```

List all pairs of plants that can grow alongside 

```python
# Define a threshold for suitability
suitability_threshold = 0.2  # You can adjust this threshold based on your requirements

# Initialize a list to store pairs of plants that exceed the threshold
exceed_threshold_pairs = []

# Iterate through each pair of plants in the suitability matrix
for i, plant1 in enumerate(plants):
    for j, plant2 in enumerate(plants):
        # Exclude diagonal cells and cells above the diagonal
        if i < j:
            suitability_value = suitability_matrix[i, j]
            # print(f"Suitibality between {plant1} and {plant2}: {suitability_value}")
            if suitability_value > suitability_threshold:
                exceed_threshold_pairs.append((plant1, plant2))

# Print the list of pairs that exceed the threshold
print("Pairs of plants that can grow alongside  :")
for pair in exceed_threshold_pairs:
    print(pair)
```
